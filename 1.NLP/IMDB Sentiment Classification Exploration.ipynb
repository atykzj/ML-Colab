# Explore different models in Sentiment Classification of IMDB Film Reviews

```python
import keras
from keras.datasets import imdb
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout
from keras.layers import Embedding 
from keras.callbacks import ModelCheckpoint  
import os 
from sklearn.metrics import roc_auc_score, roc_curve 
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

### Set hyperparameters

```python
# output directory name to store model's parameters:
output_dir = 'model_output/dense'

# training:
epochs = 4
batch_size = 128

# vector-space embedding: 
n_dim = 64 # dimensions in word-vector space
n_unique_words = 5000 # Sort all counts of word token and keep only top 5000
n_words_to_skip = 50  # 50 most frequent stop words 
max_review_length = 100 # input data is set max 100 words length, shorter are padded with characters
pad_type = trunc_type = 'pre' # trunc/add pad at start of document, content at end are more influential

# neural network architecture: 
n_dense = 64
dropout = 0.5
```

### EDA


```python
(x_train, y_train), (x_valid, y_valid) = imdb.load_data(num_words=n_unique_words, 
                                                        skip_top=n_words_to_skip) 
```

### Data Preparation for Modelling

```python
(all_x_train,_),(all_x_valid,_) = imdb.load_data() 
```

```python
' '.join(index_word[id] for id in all_x_train[0])
```

**Preprocess Data**

```python
# Stanadardising input length by padding and truncating
x_train = pad_sequences(x_train, maxlen=max_review_length, 
                        padding=pad_type, truncating=trunc_type, value=0)
x_valid = pad_sequences(x_valid, maxlen=max_review_length, 
                        padding=pad_type, truncating=trunc_type, value=0)
```

### Basic Dense Model

```python
model = Sequential()
model.add(Embedding(n_unique_words, n_dim, input_length=max_review_length))
model.add(Flatten())
model.add(Dense(n_dense, activation='relu'))
model.add(Dropout(dropout))
# model.add(Dense(n_dense, activation='relu'))
# model.add(Dropout(dropout))
model.add(Dense(1, activation='sigmoid')) # mathematically equivalent to softmax with two classes
```

```python
model.summary()
```

```python
# embedding layer dimensions and parameters: 
n_dim, n_unique_words, n_dim*n_unique_words
```

```python
# ...flatten:
max_review_length, n_dim, n_dim*max_review_length
```

```python
# ...dense:
n_dense, n_dim*max_review_length*n_dense + n_dense # weights + biases
```

```python
# ...and output:
n_dense + 1 
```

**Configure model**

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
modelcheckpoint = ModelCheckpoint(filepath=output_dir+
                                  "/weights.{epoch:02d}.hdf5")
```

```python
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

**Train**

```python
model.fit(x_train, y_train, 
          batch_size=batch_size, epochs=epochs, verbose=1, 
          validation_data=(x_valid, y_valid), 
          callbacks=[modelcheckpoint])
```

**Evaluate**

```python
# Calling saved weights
model.load_weights(output_dir+"/weights.02.hdf5") # NOT zero-indexed
```

```python
y_hat = model.predict(x_valid) 
```
